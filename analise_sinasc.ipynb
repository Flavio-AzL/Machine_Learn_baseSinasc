# üß¨ An√°lise dos Determinantes do Baixo Peso ao Nascer
## 1. Defini√ß√£o do Problema e Vari√°vel Alvo (Target)
O baixo peso ao nascer (BPN), definido pela Organiza√ß√£o Mundial da Sa√∫de (OMS) como peso inferior a **2.500g**, √© um dos principais indicadores de sa√∫de p√∫blica e preditor de mortalidade infantil. 

Para este estudo, definimos nossa vari√°vel *target* bin√°ria `baixo_peso` baseada na coluna `PESO`:
* **0 (N√£o):** Peso $\geq$ 2.500g
* **1 (Sim):** Peso < 2.500g

## 2. Sele√ß√£o de Vari√°veis (Feature Selection) Baseada em Evid√™ncias
A sele√ß√£o de atributos n√£o foi arbitr√°ria, mas fundamentada na literatura epidemiol√≥gica sobre os determinantes do BPN. O modelo considera tr√™s dimens√µes de risco:

### A. Fatores Maternos e Socioecon√¥micos
* **Idade Materna (`IDADEMAE`):** A literatura descreve uma associa√ß√£o em "Curva em U", onde adolescentes (<20 anos) e mulheres com idade avan√ßada ($\geq$35 anos) apresentam maior risco de desfechos negativos.
* **Escolaridade (`ESCMAE2010`):** Utilizada como *proxy* para o n√≠vel socioecon√¥mico. A baixa escolaridade materna est√° associada a menor acesso a informa√ß√µes de sa√∫de e nutri√ß√£o inadequada, aumentando a chance de BPN.
* **Ra√ßa/Cor (`RACACOR`):** No Brasil, estudos apontam disparidades significativas no acesso ao pr√©-natal e nos desfechos neonatais baseados na ra√ßa/cor autodeclarada, refletindo desigualdades sociais estruturais.

### B. Assist√™ncia Pr√©-Natal
* **N√∫mero de Consultas (`CONSPRENAT`):** O Minist√©rio da Sa√∫de preconiza o m√≠nimo de 6 consultas. A aus√™ncia ou baixo n√∫mero de consultas √© um fator de risco modific√°vel cr√≠tico, pois impede a detec√ß√£o precoce de condi√ß√µes como hipertens√£o gestacional.



### C. Fatores Biol√≥gicos e Obst√©tricos
* **Idade Gestacional (`SEMAGESTAC`):** A prematuridade √© o determinante proximal mais forte do baixo peso. A vari√°vel permite distinguir entre o BPN causado por prematuridade e aquele causado por restri√ß√£o de crescimento intrauterino.
* **Hist√≥rico Reprodutivo (`QTDFILVIVO`):** A paridade e o hist√≥rico de nascidos vivos anteriores influenciam o risco da gesta√ß√£o atual.
* **Sexo do Rec√©m-nascido (`SEXO`):** Vari√°vel de controle biol√≥gico, visto que rec√©m-nascidos do sexo feminino tendem a ser fisiologicamente mais leves que os do sexo masculino.
* **Anomalia Cong√™nita (`IDANOMAL`):** A presen√ßa de malforma√ß√µes √© uma causa biol√≥gica direta de restri√ß√£o de crescimento e baixo peso.

## 3. Crit√©rios de Exclus√£o e Preven√ß√£o de *Data Leakage*
Para garantir a integridade do modelo preditivo, removemos vari√°veis que configuram "vazamento de dados" (informa√ß√µes obtidas apenas *ap√≥s* o desfecho):
* **√çndice de Apgar (`APGAR1`, `APGAR5`):** Medidos no 1¬∫ e 5¬∫ minuto de vida. Sua inclus√£o enviesaria o modelo, pois s√£o consequ√™ncias do nascimento, n√£o preditores pr√©vios.
* **Dados Administrativos:** Colunas como `ORIGEM`, `CODESTAB` e `NUMEROLOTE` n√£o possuem plausibilidade biol√≥gica para influenciar o peso fetal.

#  Diagn√≥stico Estat√≠stico Inicial (Pr√©-Tratamento)

A an√°lise descritiva preliminar revelou inconsist√™ncias nos dados brutos que exigem tratamento sanit√°rio antes da modelagem. Abaixo detalhamos os problemas identificados atrav√©s das medidas de tend√™ncia central e dispers√£o:

## 1. Codifica√ß√£o de Dados Faltantes como Inteiros ("Problema do 99")
O sistema DATASUS utiliza historicamente o c√≥digo `99` (ou `9`) para representar campos "Ignorados". O `.describe()` capturou esses c√≥digos como valores num√©ricos v√°lidos, distorcendo a realidade:

* **Idade Materna (`idade_mae`):** O valor m√°ximo encontrado foi **99 anos**, biologicamente implaus√≠vel. Isso indica que registros sem informa√ß√£o de idade foram computados, elevando artificialmente a m√©dia e o desvio padr√£o.
* **Hist√≥rico Reprodutivo (`filhos_vivos`, `filhos_mortos`):** Ambos apresentam valor m√°ximo de **99**, o que √© imposs√≠vel.
* **Consultas Pr√©-Natal (`consultas_prenatal`):** O m√°ximo de **99 consultas** confirma a presen√ßa do c√≥digo de erro nesta vari√°vel cr√≠tica.

**Impacto no Modelo:** Se n√£o tratados, esses valores funcionar√£o como *outliers* extremos, impedindo o algoritmo de aprender padr√µes reais (ex: o modelo achar√° que mulheres de 99 anos t√™m muitos filhos).

## 2. Outliers e Valores Extremos no Target (`peso`)
* **M√≠nimo de 100g:** Identificamos rec√©m-nascidos com peso registrado de 100g. Pela literatura m√©dica, esse peso √© incompat√≠vel com a vida extrauterina (prov√°veis abortos tardios registrados erroneamente como nascidos vivos ou erro de digita√ß√£o).
* **M√°ximo de 6.854g:** Embora raro, a macrossomia extrema √© biologicamente poss√≠vel, mas deve ser investigada pois pode representar um ponto fora da curva (outlier) que desbalanceia a regress√£o.

## 3. Plano de Limpeza de Dados (Data Cleaning)
Com base neste diagn√≥stico, as seguintes a√ß√µes ser√£o tomadas na pr√≥xima etapa:
1.  **Substitui√ß√£o:** Converter todos os valores `99` nas colunas de idade, filhos e consultas para `NaN` (Not a Number).
2.  **Filtragem de Viabilidade:** Remover registros com `peso < 500g` (padr√£o da OMS para viabilidade fetal), pois distorcem a an√°lise de baixo peso.

#  Tratamento e Limpeza dos Dados (Data Cleaning)
Baseado no diagn√≥stico estat√≠stico anterior, aplicaremos as regras de limpeza para garantir a qualidade do modelo.

```python
import numpy as np

# 1. Tratamento do "Erro 99" (Substituir 99 por NaN)
# Colunas onde 99 significa "Ignorado" segundo o manual
cols_correcao_99 = ['idade_mae', 'filhos_vivos', 'filhos_mortos', 'consultas_prenatal']

# Loop para substituir
for col in cols_correcao_99:
    # Apenas se a coluna existir no dataframe
    if col in df_analise.columns:
        # Substitui 99 por NaN
        df_analise[col] = df_analise[col].replace(99, np.nan)

# 2. Tratamento de Viabilidade (Peso)
# Remover beb√™s com peso < 500g (considerados aborto/invi√°veis para an√°lise de baixo peso comum)
# Tamb√©m removemos pesos vazios (NaN) pois √© nosso Target
df_limpo = df_analise[
    (df_analise['peso'] >= 500) & 
    (df_analise['peso'].notnull())
].copy() # .copy() cria um novo dataframe independente na mem√≥ria

# 3. Tratamento da Escolaridade e Ra√ßa (Onde 9 ou 0 costumam ser "Ignorado")
# Nota: Verifique se na sua base 'raca_cor' vai de 1 a 5. Se tiver 9, √© ignorado.
if 'raca_cor' in df_limpo.columns:
    df_limpo['raca_cor'] = df_limpo['raca_cor'].replace(9, np.nan)

# ---------------------------------------------------------
# VERIFICA√á√ÉO P√ìS-TRATAMENTO
# ---------------------------------------------------------
print(f"Linhas antes: {len(df_analise)}")
print(f"Linhas depois: {len(df_limpo)}")
print(f"Perda de dados: {100 - (len(df_limpo)/len(df_analise)*100):.2f}%")

print("\nüìä Estat√≠sticas dos Dados Limpos:")
display(df_limpo.describe())
```

```python
# Ainda ficaram outliers, o 9 est√° sendo considerado como ignorado para a escolaridade da M√£e
# O n√∫mero 0 esta sendo considerado para o sexo ignorado tamb√©m


# 1. Desligar a nota√ß√£o cient√≠fica (e+01, e+06)
# Agora os n√∫meros v√£o aparecer com 2 casas decimais normais (ex: 9.00)
pd.options.display.float_format = '{:.2f}'.format

# 2. REFINANDO A LIMPEZA (Rodada 2)
# Vamos limpar o que encontramos na nova inspe√ß√£o

# Lista de colunas onde '9' significa "Ignorado" (Escolaridade e Ra√ßa)
cols_remove_9 = ['escolaridade_mae', 'raca_cor', 'anomalia_congenita']

for col in cols_remove_9:
    if col in df_limpo.columns:
        df_limpo[col] = df_limpo[col].replace(9, np.nan)

# Lista de colunas onde '0' significa "Ignorado" (Sexo costuma ter 0 como ignorado)
if 'sexo' in df_limpo.columns:
    # Remove sexo 0 (Ignorado) e 9 (se houver)
    df_limpo['sexo'] = df_limpo['sexo'].replace([0, 9], np.nan)

# EXTRA: Capando valores absurdos (Outliers extremos de digita√ß√£o)
# Exemplo: 30 filhos ou 80 consultas √© tecnicamente poss√≠vel, mas estatisticamente prov√°vel erro
df_limpo.loc[df_limpo['consultas_prenatal'] > 40, 'consultas_prenatal'] = np.nan
df_limpo.loc[df_limpo['filhos_vivos'] > 20, 'filhos_vivos'] = np.nan

# 3. VERIFICANDO DE NOVO
print("Tabela Limpa (Sem nota√ß√£o cient√≠fica):")
display(df_limpo.describe())

# 4. "TRADU√á√ÉO" DOS DADOS CATEG√ìRICOS
# Para voc√™ visualizar o que os n√∫meros significam
print("\n--- Dicion√°rio de Dados (O que os n√∫meros significam) ---")
print("Sexo: 1=Masc, 2=Fem")
print("Ra√ßa: 1=Branca, 2=Preta, 3=Amarela, 4=Parda, 5=Ind√≠gena")
print("Escolaridade: 0=Sem, 1=Fund I, 2=Fund II, 3=M√©dio, 4=Sup Inc, 5=Sup Comp")
```

# A Regra Cient√≠fica (Hierarquia de Decis√£o)
## N√£o, nem tudo deve ser exclu√≠do!

Transformar o dado "sujo" (99, 9) em NaN √© apenas o primeiro passo. O segundo passo √© decidir o que fazer com esses NaNs. Se voc√™ excluir todas as linhas que t√™m algum NaN, voc√™ pode jogar fora 30% ou 40% da sua base sem necessidade.

Para decidir entre Excluir vs. Preencher (Imputar), n√≥s usamos uma "m√©trica" chamada Taxa de Dados Faltantes (Missing Rate).

Aqui est√° a regra de ouro cient√≠fica para o trabalho:
Voc√™ deve calcular a % de nulos em cada coluna e aplicar a regra:

1. Se for o TARGET (peso): EXCLUIR (Drop)
Motivo: Voc√™ nunca deve inventar (imputar) o valor que voc√™ quer prever. Se o beb√™ n√£o tem peso registrado, ele n√£o serve para treinar o modelo.

A√ß√£o: dropna nas linhas onde peso √© NaN.

2. Se for uma Feature (Idade, Escolaridade) com POUCOS nulos (< 5%): EXCLUIR (Drop)
Motivo: Como sua base √© gigante (2.5 milh√µes de linhas), perder 1% ou 2% das linhas n√£o afeta a estat√≠stica e √© mais seguro do que inventar dados.

A√ß√£o: dropna.

3. Se for uma Feature com MUITOS nulos (5% a 30%): PREENCHER (Imputar)
Motivo: Se voc√™ apagar todas as m√£es que n√£o informaram escolaridade (que costumam ser muitas), voc√™ pode introduzir um vi√©s. Talvez as m√£es mais pobres sejam as que menos informam escolaridade. Se voc√™ apagar essas linhas, seu modelo vai "aprender" apenas sobre m√£es ricas.

A√ß√£o: Preencher com a Mediana (num√©ricos) ou Moda (categ√≥ricos).

4. Se for uma Feature com NULOS DEMAIS (> 50%): DESCARTAR A COLUNA
Motivo: Se metade da informa√ß√£o n√£o existe, a vari√°vel √© in√∫til.

A√ß√£o: drop na coluna inteira.

## DIAGN√ìSTICO DE DADOS FALTANTES (MISSING DATA AUDIT)

```python

# Calcula a porcentagem de NaN em cada coluna
percentual_nulos = (df_limpo.isnull().sum() / len(df_limpo)) * 100

# Cria uma tabela bonita para decis√£o
tabela_decisao = pd.DataFrame({
    'Colunas': percentual_nulos.index,
    '% Nulos': percentual_nulos.values,
    'Decis√£o Sugerida': [
        'Excluir Linhas (Target)' if col == 'peso' else
        'Excluir Linhas (Poucos Nulos)' if 0 < x < 5 else
        'Preencher (Imputa√ß√£o)' if 5 <= x < 30 else
        'Descartar Coluna' if x >= 30 else
        'Ok (Completo)'
        for col, x in zip(percentual_nulos.index, percentual_nulos.values)
    ]
})

print("üìâ Relat√≥rio de Estrat√©gia para Dados Faltantes:")
display(tabela_decisao.sort_values('% Nulos', ascending=False))
```

## Aplicando a corre√ß√µes para a limpeza

```python
# 1. TARGET: Se n√£o tem peso, drop!
df_final = df_limpo.dropna(subset=['peso']).copy()

# ---------------------------------------------------------
# 2. TRATAMENTO DE OUTLIERS (ANTES DE IMPUTAR)
# ---------------------------------------------------------
# Regra: Valores acima disso viram NaN para serem tratados pela mediana depois
limite_filhos_vivos = 12
limite_filhos_mortos = 10

# Aplica√ß√£o: Transforma os excessos em NaN
df_final.loc[df_final['filhos_vivos'] > limite_filhos_vivos, 'filhos_vivos'] = np.nan
df_final.loc[df_final['filhos_mortos'] > limite_filhos_mortos, 'filhos_mortos'] = np.nan

# Limpeza extra de outliers de digita√ß√£o (Consultas > 40 viram NaN)
df_final.loc[df_final['consultas_prenatal'] > 40, 'consultas_prenatal'] = np.nan

# ---------------------------------------------------------
# 3. IMPUTA√á√ÉO NUM√âRICA 
# ---------------------------------------------------------
# Agora que os outliers viraram NaN, a mediana vai preencher eles tamb√©m
cols_imputar_mediana = ['idade_mae', 'filhos_vivos', 'filhos_mortos', 'semanas_gestacao', 'consultas_prenatal']

for col in cols_imputar_mediana:
    if col in df_final.columns:
        mediana = df_final[col].median()
        df_final[col] = df_final[col].fillna(mediana)

# ---------------------------------------------------------
# 4. IMPUTA√á√ÉO CATEG√ìRICA
# ---------------------------------------------------------
cols_imputar_moda = ['escolaridade_mae', 'sexo', 'raca_cor', 'anomalia_congenita']

for col in cols_imputar_moda:
    if col in df_final.columns:
        moda = df_final[col].mode()[0]
        df_final[col] = df_final[col].fillna(moda)

# ---------------------------------------------------------
# 5. VERIFICA√á√ÉO FINAL
# ---------------------------------------------------------
nulos = df_final.isnull().sum().sum()
print(f"Nulos restantes: {nulos} (deve ser ZERO)")
print(f"Registros finais: {len(df_final)}")

# Check visual
display(df_final.describe())
```
